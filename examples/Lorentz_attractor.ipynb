Lorentz attractor の課題解答例

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def lorentz_next(x, y, z, dt, a, b, mu):
    """ Get a next step with given dt, a, b, mu """
    x_new = x + (- a * x + a * y) * dt
    y_new = y + (mu * x - y - x * z) * dt
    z_new = z + (- b * z + x * y) * dt
    return x_new, y_new, z_new
```

## With variable $\mu$

```python
a = 10.0
b = 8.0/3.0
```

```python
def lorentz_simulation(x0, y0, z0, dt, a, b, mu, n):
    x = [x0]
    y = [y0]
    z = [z0]
    for _ in range(n):
        x_new, y_new, z_new = lorentz_next(x[-1], y[-1], z[-1], 
                                           dt, a, b, mu)
        x.append(x_new)
        y.append(y_new)
        z.append(z_new)
    return x, y, z
```

```python
x0 = 10.0
y0 = 10.0
z0 = 0.1
# mu=0
mu = 0
x_mu0, y_mu0, z_mu0 = lorentz_simulation(x0, y0, z0, 0.01, a, b, 
                                         mu, 10000)
# mu=1
mu = 1.0
x_mu1, y_mu1, z_mu1 = lorentz_simulation(x0, y0, z0, 0.01, a, b, 
                                         mu, 10000)
# mu=3
mu = 3.0
x_mu3, y_mu3, z_mu3 = lorentz_simulation(x0, y0, z0, 0.01, a, b, 
                                         mu, 10000)
# mu=10
mu = 10.0
x_mu10, y_mu10, z_mu10 = lorentz_simulation(x0, y0, z0, 0.01, a, b, 
                                         mu, 10000)
# mu=30
mu = 30.0
x_mu30, y_mu30, z_mu30 = lorentz_simulation(x0, y0, z0, 0.01, a, b, 
                                         mu, 10000)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.figure(figsize=(10,10))
ax = plt.subplot(111, projection='3d')
ax.plot(x_mu0, y_mu0, z_mu0)
ax.plot(x_mu1, y_mu1, z_mu1)
ax.plot(x_mu3, y_mu3, z_mu3)
ax.plot(x_mu10, y_mu10, z_mu10)
ax.plot(x_mu30, y_mu30, z_mu30)
```
